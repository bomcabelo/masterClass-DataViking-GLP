```python
def Validar_Arquivo( Base_Dados, Nome_Arquivo):

  '''
  Funo ir verifiicar algumas estatisticas da ingesto dos dados
  '''

  Linhas = Base_Dados.shape[0]
  Colunas = Base_Dados.shape[1]

  Valores_Nulos = Base_Dados.isnull().sum()
  Total_Nulos = Valores_Nulos.sum()
  Colunas_Nulas = Valores_Nulos[ Valores_Nulos > 0  ].index

  Tipos_Campos = Base_Dados.dtypes

  Regioes = Base_Dados['Regiao - Sigla'].unique()
  Estados = Base_Dados['Estado - Sigla'].unique()

  Nome_Arquivo = str(Nome_Arquivo)
  Data = datetime.datetime.now().date()
  Hora = datetime.datetime.now().time()

  Dicionario = {
    'Total de Registros' : Linhas,
    'Totla de Colunas' : Colunas,
    'Total_Linhas_Nulas' : Total_Nulos,
     'Colunas_Nulas' : [ Colunas_Nulas ],
     'Tipo_Campos' : [Tipos_Campos],
    'Estatistica_Regiao' : [Regioes],
    'Estatistica_Estados' : [Estados],
    'Nome_Arquivo' : Nome_Arquivo,
    'Data' : Data,
    'Hora' : Hora
  }

  Log = pd.DataFrame( Dicionario, index=[0] )

  return Log
```

```python
# Estrutura para consolidar os arquivo Baixados

# Local para Buscar os arquivos Baixados
Local = '/content/'

# Variaveiis de apoio
Base_Consolidada = pd.DataFrame()
Base_Log = pd.DataFrame()

# Loop para percorrer o diretorio
for Diretorio, Subpastas, Arquivos in os.walk( Local ):

  #print(Diretorio, Subpastas, Arquivos)
  for Arquivo in Arquivos:

    if '.csv' in Arquivo and 'glp' in Arquivo:

      # Empilhando os dados
      Base = pd.read_csv( Local + f'{Arquivo}', sep=';' )
      Base_Consolidada = pd.concat( [Base_Consolidada, Base] )

      # Log
      Base_Log = pd.concat([Base_Log, Validar_Arquivo(Base, Arquivo ) ])

    else:
      pass
```

**Desvio Padro** <br>
<p>
O desvio padro  uma medida estatstica que quantifica a disperso ou variabilidade dos dados em um conjunto de dados.
</p>
<p>
Ele indica o quanto os valores em um conjunto de dados esto dispersos ou afastados da mdia (valor mdio). Em outras palavras, o desvio padro fornece uma medida de quo "espalhados" os dados esto ao redor da mdia.
</p>

**Assimetria**

image.png

**Em termos prticos:**

1. Quando a **assimetria  igual a 0**, a distribuio  considerada simtrica, o que significa que os valores esto igualmente distribudos ao redor da mdia, e no h inclinao para a direita ou para a esquerda.

2. Uma **assimetria negativa (menor que 0)** indica que a cauda da distribuio est mais longa  esquerda em comparao com a direita. Em outras palavras, a distribuio se estende mais para valores menores do que para valores maiores em relao  mdia.

3. Uma **assimetria positiva (maior que 0)** indicaria o oposto, ou seja, que a cauda da distribuio estaria mais longa  direita em comparao com a esquerda.

_______________________

Os benefcios de ter uma distribuio normal ou aproximadamente normal incluem:

1. **Facilidade de anlise estatstica**: Muitos mtodos estatsticos e testes de hipteses assumem que os dados seguem uma distribuio normal. Quando os dados se aproximam dessa distribuio, esses mtodos podem ser aplicados com mais confiana e preciso.

2. **Interpretao intuitiva**: Distribuies normais so bem compreendidas e tm propriedades matemticas bem estabelecidas, como a regra emprica (68-95-99.7), que descreve a proporo de dados dentro de determinados intervalos de desvio padro da mdia.

3. **Facilita a previso**: Muitos modelos estatsticos e de machine learning assumem que os erros seguem uma distribuio normal. Quando os erros so normalmente distribudos, os modelos podem fazer previses mais precisas.

```python
# Criar um histograma
plt.title('Analise preo de venda')

# Plot
plt.hist( Base_Dados['valor_de_venda'], bins=10, density=True, alpha=0.5, color='b')

# Adicionar um media
Media_Preco = Base_Dados['valor_de_venda'].mean()
plt.axvline( Media_Preco, color='red', linestyle='dashed', linewidth=2, label='mdia' )

# Assimetria
Assimetria = Base_Dados['valor_de_venda'].skew()
plt.annotate( f'Assimetria = {Assimetria:.2f}', xy=(0.75, 0.95), xycoords='axes fraction', fontsize=9 )

plt.xlabel('Valores')
plt.ylabel('Frequncia');
```

```python
# Serie temporal
Anl_Serie = Base_Dados.groupby(
     by=['ano_coleto', 'mes_coleto']
).agg(
    Media = ('valor_de_venda', 'mean')
).reset_index()

Anl_Serie.head()
```

```python
Anl_Serie['Periodo'] = Anl_Serie['ano_coleto'].astype('str') + '-' + Anl_Serie['mes_coleto'].astype('str') + '-' + '01'
Anl_Serie['Periodo'] =  pd.to_datetime( Anl_Serie['Periodo'] )
Anl_Serie.set_index('Periodo', inplace=True )

Anl_Serie.head()
```

```python
# Gestoo do Ministro da Fazendo
Anl_Serie['Gesto'] = Anl_Serie['ano_coleto'].apply( lambda Loop: 'Paulo Guedes' if Loop < 2023 else 'Fernando Haddad' )
Anl_Serie.head()
```

```python
Anl_Serie.tail()
```

```python
Anl_Serie['Oscilao'] = (( Anl_Serie.Media / Anl_Serie.Media.shift(1) ) - 1) * 100

Anl_Serie.head()
```

```python
# Grfico de disperso + Linha da Regresso + r

# Libs
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Treinar uma regresso
x = Anl_Serie['Indice_Brent'].values.reshape(-1, 1)
y = Anl_Serie['Media'].values.reshape(-1, 1)

# Treinao da Regresso
Regressor = LinearRegression()
Regressor.fit(x, y)

# Calcule o coeficiente de determinao (R)
y_pred = Regressor.predict(x)
r2 = r2_score(y, y_pred)

# Tamanho da Imagem
plt.figure(figsize=(8, 6))

# Plot
sns.regplot(
    data=Anl_Serie, x='Media', y='Indice_Brent', color='black',
    scatter_kws={'s': 50}, line_kws={'color': 'red'} )

plt.title(f'Scatter Plot com Linha de Regresso (R = {r2:.2f})')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid(True)
```
